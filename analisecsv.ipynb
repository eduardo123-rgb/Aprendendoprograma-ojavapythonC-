```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('2023_Pagamento.csv', encoding='cp1252', sep=';')
print(dados)
dados['Valor'] = dados['Valor'].str.replace(',','.')
dados['Valor'] = dados['Valor'].astype(float)
```

```python
edua = dados[dados['Nome do órgão superior']=='Ministério da Defesa']
print(edua)
```

```python
total= edua['Valor'].sum()
media = np.mean(edua['Valor'])
desvio = np.std(edua['Valor'])
maximovalor = np.max(edua['Valor'])
minimovalor = np.min(edua['Valor'])
```

```python
print('A media em real gasto pelo ministerio da defesa neste ano é ',media)
print('O desvio desta medai gasta pelo ministerio da defesa neste ano é ',desvio)
print('O maior valor usado é ',maximovalor,'Real')
print('O minimo valor gasto é ',minimovalor,'Real')
print('E o total gasto pelo ministerio da defesa é ',total,'Real')
```

```python
d = dados['Valor'].sum()
print(d)
```

```python
percentage = (total/d)*100
print(percentage,'%')
```

```python
ed= edua[edua['Tipo de pagamento']=='DIÁRIAS']
print(ed)
```

```python
pga = ed['Valor'].sum()
print('Valor gasto com diárias ',pga)
per = (pga/total)*100
print('Percentagem do gasto em diaria do gasto total do Ministerio da Defesa',per,'%')
```

```python
eduaq = edua[edua['Tipo de pagamento']=='PASSAGEM']
print(eduaq)
```

```python
po = eduaq['Valor'].sum()
print(po)
prq = (po/total)*100
print(prq,'%')
```

```python
eduau = edua[edua['Tipo de pagamento']=='Serviço correlato: seguro']
print(eduau)
uy = eduau['Valor'].sum()
```

```python
pert = (uy/total)*100
print(pert)
```

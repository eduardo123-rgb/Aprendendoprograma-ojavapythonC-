```python
#importando os pacotes necessario
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Adicionando o dataset no meu projecto
#E transformando os dado numérico da coluna Valor de string para tipo float para ser possível fazer cálculos 
dados = pd.read_csv('2023_Pagamento.csv', encoding='cp1252', sep=';')
dados['Valor'] = dados['Valor'].str.replace(',','.')
dados['Valor'] = dados['Valor'].astype(float)
```

```python
#filtrando apenas o ministerio da defesa para ser analisado posteriormente vamos analisar outros !
edua = dados[dados['Nome do órgão superior']=='Ministério da Defesa']
```

```python
edua.to_csv('MiniserioDEF.csv',index=False)
```

```python
#calculo de total em dinheiro de toda dataset
d = dados['Valor'].sum()
print(f' O total Valor em real da dataset {d} real')
```

```python
#dados estatistico e total gasto pelo ministerio da defesa !
total= edua['Valor'].sum()
media = np.mean(edua['Valor'])
desvio_padrao = np.std(edua['Valor'])
maximovalor = np.max(edua['Valor'])
minimovalor = np.min(edua['Valor'])
percentage = (total/d)*100
print(f'O ministerio da defesa gastou {percentage}% do total em dinheiro da dataset')
print(f'A media em real gasto pelo ministerio da defesa neste ano é {media}')
print(f'O desvio padrão desta media gasta pelo ministerio da defesa neste ano é {desvio_padrao}')
print(f'O maior valor usado é {maximovalor}Real')
print(f'O minimo valor gasto é {minimovalor}Real')
print(f'E o total gasto pelo ministerio da defesa é {total}Real')
```
